### Anthropic SDKã¨Messages API
ã“ã®ãƒãƒ¥ãƒ¼ãƒˆãƒªã‚¢ãƒ«ã§ã¯ã€[Anthropic Python SDK](https://docs.anthropic.com/claude/reference/claude-on-amazon-bedrock)ã¨[Messages API](https://docs.anthropic.com/claude/reference/messages_post)ã‚’ä½¿ç”¨ã—ã¾ã™ã€‚

ä»¥ä¸‹ã§ã¯ã€ã“ã®ãƒãƒ¥ãƒ¼ãƒˆãƒªã‚¢ãƒ«ã§Claudeã«ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã‚’é€ä¿¡ã™ã‚‹éš›ã®å®Ÿä¾‹ã‚’ç¤ºã—ã¾ã™ã€‚

ã¾ãšã€ä½¿ç”¨ã™ã‚‹ãƒ¢ãƒ‡ãƒ«åã¨AWSãƒªãƒ¼ã‚¸ãƒ§ãƒ³ã‚’è¨­å®šã—ã€ä¿å­˜ã—ã¾ã™ã€‚

```python
import boto3
session = boto3.Session() # boto3ã‚»ãƒƒã‚·ãƒ§ãƒ³ã‚’ä½œæˆã—ã¦ãƒªãƒ¼ã‚¸ãƒ§ãƒ³åã‚’å‹•çš„ã«å–å¾—
AWS_REGION = session.region_name
print("AWS Region:", AWS_REGION)
MODEL_NAME = "anthropic.claude-3-haiku-20240307-v1:0"

%store MODEL_NAME
%store AWS_REGION
```

æ¬¡ã«ã€`get_completion`é–¢æ•°ã‚’å®šç¾©ã—ã¾ã™ã€‚ã“ã®é–¢æ•°ã¯ã€ä¸ãˆã‚‰ã‚ŒãŸãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã‚’Claudeã«é€ä¿¡ã—ã€ç”Ÿæˆã•ã‚ŒãŸå¿œç­”ã‚’è¿”ã™ãƒ˜ãƒ«ãƒ‘ãƒ¼é–¢æ•°ã§ã™ã€‚ä»¥ä¸‹ã®ã‚»ãƒ«ã‚’å®Ÿè¡Œã—ã¦ãã ã•ã„ã€‚

```python
from anthropic import AnthropicBedrock

client = AnthropicBedrock(aws_region=AWS_REGION)

def get_completion(prompt, system=''):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        messages=[
          {"role": "user", "content": prompt}
        ],
        system=system
    )
    return message.content[0].text
```

ã§ã¯ã€å®Ÿéš›ã«Claudeã«å¯¾ã—ã¦ã‚µãƒ³ãƒ—ãƒ«ã®ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã‚’é€ä¿¡ã—ã€`get_completion`é–¢æ•°ã‚’ä½¿ã£ã¦å¿œç­”ã‚’å–å¾—ã—ã¦ã¿ã¾ã—ã‚‡ã†ã€‚ä»¥ä¸‹ã®ã‚»ãƒ«ã‚’å®Ÿè¡Œã™ã‚‹ã¨ã€ãã®ä¸‹ã«Claudeã‹ã‚‰ã®å¿œç­”ãŒè¡¨ç¤ºã•ã‚Œã¾ã™ã€‚

ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã®å†…å®¹ã‚’è‡ªç”±ã«å¤‰æ›´ã—ã¦ã€æ§˜ã€…ãªå¿œç­”ã‚’è©¦ã—ã¦ã¿ã‚‹ã“ã¨ãŒã§ãã¾ã™ã€‚

```python
# ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆ
prompt = "ã“ã‚“ã«ã¡ã¯ã€Claude!"

# Claudeã‹ã‚‰ã®å¿œç­”ã‚’å–å¾—
print(get_completion(prompt))
```
